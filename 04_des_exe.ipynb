```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Exercise 4.1

Implementing Equation 4.1 
$$f(x) = \frac{1}{\sigma \sqrt{2\pi}}e^{\frac{-x^2}{2\sigma^2}}$$

```python
# Definining the distribution parameters

sigma = 0.73
pi = np.pi
x = np.linspace(-3,3, 200)
```

```python
# Breaking down the equation

mult = 1 / (sigma * (2 * pi)**1/2)
eterm = -x**2 / (2 * sigma**2)

eqn = mult * np.exp(eterm)
```

```python
fig, ax = plt.subplots(figsize=(4, 3))

ax.plot(x, eqn, linewidth=2, color='black')
ax.set_xlabel('x')
ax.set_ylabel('Height')
ax.set_title('A)', loc='left')
```

Group of Gaussians

```python
N = 50
sigmas = np.linspace(0.1,3,N)
```

```python
G = np.zeros((N, len(x)))

for i in range(N):
    mult = 1 / (sigmas[i]*np.sqrt(2*pi))
    eterm = -x**2 / (2*sigmas[i]**2)
    G[i,:] = mult * np.exp(eterm)
```

```python
fewGaussians = np.round(np.linspace(4, N-1, 8)).astype(int)
```

```python
fig, ax = plt.subplots(figsize=(5,4))

ax.plot(G[fewGaussians, :].T)
ax.set_title('B)', loc='left')
ax.set_xlabel('x')
# Can't get the xlim right
ax.set_ylabel('height')
ax.legend([f"$\sigma = {s:.2f}$" for s in sigmas[fewGaussians]])
```

Family Portrait

```python
# Test plot
plt.imshow(G,extent=[x[0],x[-1],sigmas[0],sigmas[-1]],
           origin='lower')
```

```python
# Final plot

fig, axs = plt.subplots(1,3, figsize=(13,4))

axs[0].plot(x, eqn, linewidth=2, color='black')
axs[0].set_xlabel('x')
axs[0].set_ylabel('Height')
axs[0].set_xlim(x[[0,-1]])
axs[0].set_title('A)', loc='left')

axs[1].plot(G[fewGaussians, :].T)
axs[1].set_title('B)', loc='left')
axs[1].set_xlabel('x')
#axs[1].set_xlim(x[[0,-1]])
# why isn't this working?
axs[1].set_ylabel('Height')
axs[1].legend([f"$\sigma = {s:.2f}$" for s in sigmas[fewGaussians]])

axs[2].imshow(G,extent=[x[0],x[-1],sigmas[0],sigmas[-1]],
           cmap='gray',aspect='auto',origin='lower',
           vmin=0,vmax=.8)
axs[2].set_xlabel('x')
axs[2].set_ylabel(f'$\sigma$')
axs[2].set_title('C)', loc='left')
```
