### Fig 2.3: the "6" data

Images can be stored as a matrix of numbers. Grayscale pictures are stored as a 2D matrix; color pictures are stored as 3D. 

```python
# import MNIST data
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784', as_frame=False, cache=False, parser='auto')
     
```

```python
# show one number
_,axs = plt.subplots(1,2,figsize=(8,6))

# the image of the number
I = mnist.data[18].reshape(28,28)[2:24,:][:,8:-7]
axs[0].imshow(I,cmap='gray')
axs[0].axis('off')

axs[1].imshow(I,cmap='gray',vmin=-1,vmax=0)
axs[1].axis('off')

# and the numbers of the number
for i in range(I.shape[0]):
  for j in range(I.shape[1]):
    axs[1].text(j,i,int(I[i][j]),fontsize=8
                 ,horizontalalignment='center',verticalalignment='center')

plt.tight_layout()
plt.show()
```

### Fig 2.5: Impact of noise on statistical analysis and interpretation

Noise in the data can negatively impact statistical analysis and interpretation.

```python
# generate data
n = 30
x = np.random.randn(n)
y1 = x + np.random.randn(n)/10
y2 = x + np.random.randn(n)
```

```python
# plot
_,axs = plt.subplots(2,1,figsize=(2,4))

axs[0].plot(x,np.polyval(np.polyfit(x,y1,1),x),color='gray')
axs[0].plot(x,y1,'ws',markeredgecolor='k')
axs[0].set_title('Less noise',loc='center')

axs[1].plot(x,np.polyval(np.polyfit(x,y2,1),x),color='gray')
axs[1].plot(x,y2,'ws',markeredgecolor='k')
axs[1].set_title('More noise',loc='center')

for a in axs:
  a.set_xticks([])
  a.set_xlabel('x')
  a.set_yticks([])
  a.set_ylabel('y')

# display the figure
plt.tight_layout()
plt.show()
```

### Fig 2.6: Outliers

Outliers are unusual data values. They may skew statistics and can have an outsized impact on the results. 

```python
x = np.random.randn(12)
x[6] = 2*np.pi
```

```python
# plot
plt.figure(figsize=(4,2))
plt.plot(x,'ko',markersize=10)
plt.plot(6,x[6],'ko',markersize=10,markerfacecolor=(.7,.7,.7))

# ticks and labels
plt.xticks([])
plt.yticks([])
plt.ylim([np.min(x)-.6,np.max(x)+.6])
plt.xlabel('Data index')
plt.ylabel('Data value')

# display the figure
plt.tight_layout()
plt.show()
```
