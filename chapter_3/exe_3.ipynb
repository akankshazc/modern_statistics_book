```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Data dict

d = {
    'Chocolate': 24,
     'Vanilla' : 16,
  'Strawberry' :  7,
   'Pistachio' : 16
   }
```

```python
data = pd.Series(d, index=d.keys())
data
```

```python
# Plot

fig, ax = plt.subplots(figsize=(4,4))

ax.pie(data, labels=data.index, autopct="%1.1f%%", explode=[0, 0.1, 0, 0.15]);
```
